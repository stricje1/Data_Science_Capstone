```python
#1 Importing the librariesimport numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
#2 Download Mall_Cusomers dataset from 'https://www.kaggle.com/akram24/mall-customers#Mall_Customers.csv' to local drive
#3 Importing the Mall_Customers dataset by pandas
dataset = pd.read_csv('D:/Documents/Data/Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

```python
#4 Fit a hierarchical clustering to the Mall_Customes dataset using Agglomerative Hierarchical Clustering. 
# Choose Euclidean distance and Ward method for the algorithm class 
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage ='ward')

# Try to fit the hierarchical clustering algorithm  to dataset X while creating the clusters vector
# that indicates to which cluster each customer belongs.
y_hc=hc.fit_predict(X)
```

```python
#5 Visualizing the clusters. 
# Let y_hc represent the hierarchical clustering vector
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], s=100, c='green', label ='Cluster 3')
plt.scatter(X[y_hc==3, 0], X[y_hc==3, 1], s=100, c='cyan', label ='Cluster 4')
plt.title('Clusters of Customers (Hierarchical Clustering Model)')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100')
plt.show()
```
