```python
# Implementing Ridge Regression from scratch using Python. 
# Import the following libraries:
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import Ridge

# Use the scikit-learn library to generate sample data which is well suited for regression.
X, y, coefficients = make_regression(
    n_samples=50,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=5,
    coef=True,
    random_state=1
)

# Define the hyperparameter alpha. Alpha determines the regularization strength.
alpha = 1

# Create the identity matrix.
n, m = X.shape
I = np.identity(m)

# Solve for w using the equation discuss above.
w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X) + alpha * I), X.T), y)
```

```python
# Compare w to the actual coefficient(s) used in generating the data.
w, coefficients
```

```python
# Visualize how the regression line fits the data.
plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```

```python
# Repeat using the scikit-learn implementation of Ridge Regression. 
# Create and train an instance of the Ridge class.
rr = Ridge(alpha=1)
rr.fit(X, y)
w = rr.coef_

# Get the value for w and compare it to its value linear regression.
w
```

```python
# Visualize the regression line and compare it to the one above.
plt.scatter(X, y)
plt.plot(X, w*X, c='red')
```

```python
# Visualize the effect of the regularization parameter alpha. 
# To start, set alpha to 10.

rr = Ridge(alpha=10)
rr.fit(X, y)
w = rr.coef_[0]
plt.scatter(X, y)
plt.plot(X, w*X, c='green')
```

```python
# For emphasis, try an alpha of 100.

rr = Ridge(alpha=100)
rr.fit(X, y)
w = rr.coef_[0]
plt.scatter(X, y)
plt.plot(X, w*X, c='blue')

# Expain what happens when alpa increases from 1 to 100 (or higher)
```
